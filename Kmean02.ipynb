```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('/content/xclara.csv')
```

```python
df.info()
```

```python
plt.scatter(df['V1'],df['V2'])
```

```python
from sklearn.cluster import KMeans
```

```python
wccs=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(df)
  wccs.append(kmeans.inertia_)
```

```python
wccs
```

```python
plt.plot(range(1,11),wccs)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```python
X=df.iloc[:,[0,1]].values
kmeans=KMeans(n_clusters=3,init='k-means++',random_state=42)
y_kmeans=kmeans.fit_predict(X)
```

```python
X[y_kmeans==0,1]
```

```python
plt.scatter(X[y_kmeans==0,0],X[y_kmeans==0,1],s=100,c='red',label='Cluster 1')
plt.scatter(X[y_kmeans==1,0],X[y_kmeans==1,1],s=100,c='blue',label='Cluster 2')
plt.scatter(X[y_kmeans==2,0],X[y_kmeans==2,1],s=100,c='green',label='Cluster 3')
```

```python


import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X[y_kmeans==0,0], X[y_kmeans==0,1], np.zeros_like(X[y_kmeans==0,0]), s=100, c='red', label='Cluster 1')
ax.scatter(X[y_kmeans==1,0], X[y_kmeans==1,1], np.zeros_like(X[y_kmeans==1,0]), s=100, c='blue', label='Cluster 2')
ax.scatter(X[y_kmeans==2,0], X[y_kmeans==2,1], np.zeros_like(X[y_kmeans==2,0]), s=100, c='green', label='Cluster 3')

ax.set_xlabel('V1')
ax.set_ylabel('V2')
ax.set_zlabel('Z')
plt.legend()
plt.show()
```
