```python


import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_digits
import seaborn as sn
%matplotlib inline

# ... (Previous code remains the same)

# Define a function to build and train the model with different optimizers and losses
def train_model(optimizer, loss):
    model = keras.Sequential([
        keras.layers.Flatten(input_shape=(28, 28)),
        keras.layers.Dense(100, activation='relu'),
        keras.layers.Dense(10, activation='sigmoid')
    ])
    model.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
    model.fit(x_train, y_train, epochs=5)
    _, accuracy = model.evaluate(x_test, y_test)
    y_predicted = model.predict(x_test)
    y_predicted_labels = [np.argmax(i) for i in y_predicted]
    cm = tf.math.confusion_matrix(labels=y_test, predictions=y_predicted_labels)

    return accuracy, cm


# Experiment with different optimizers and losses
optimizers = ['adam', 'sgd', 'rmsprop']
losses = ['sparse_categorical_crossentropy', 'categorical_crossentropy'] # Added categorical_crossentropy

results = {}

for optimizer in optimizers:
    for loss in losses:
      try:
        accuracy, cm = train_model(optimizer, loss)
        results[(optimizer, loss)] = (accuracy, cm)
        print(f"Optimizer: {optimizer}, Loss: {loss}, Accuracy: {accuracy}")
        plt.figure(figsize=(10, 7))
        sn.heatmap(cm, annot=True, fmt='d')
        plt.xlabel('Predicted')
        plt.ylabel('Truth')
        plt.title(f'Confusion Matrix - {optimizer} - {loss}')
        plt.show()
      except Exception as e:
        print(f"Error with optimizer {optimizer} and loss {loss}: {e}")

# Analyze results (example)
# ... (Add your analysis of the results based on the accuracy and confusion matrices)
```
