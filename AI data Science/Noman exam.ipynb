```python
## we have to create boolen series for every condtion 
#a) Low (Score < 5)
#b) Medium (5 <= Score < 7)
#c) High (Score >= 7)
# After this we update our columns on the basis of the boolen series
```

```python
whr_pd["Score"]<5   ## condition 1
```

```python
whr_pd.loc[whr_pd["Score"]<5,"Happiness Score Category"]="Low"       ## updated
```

```python
(whr_pd["Score"]>=5) & (whr_pd["Score"]<7)   ##Condition 2
```

```python
whr_pd.loc[(whr_pd["Score"]>=5) & (whr_pd["Score"]<7),"Happiness Score Category"]="Medium"  ## updated
```

```python
whr_pd["Score"]>=7   #condition Three
```

```python
whr_pd.loc[whr_pd["Score"]>=7,"Happiness Score Category"]="High"  ## updated
```

```python
whr_pd["Country or region"]=="Finland"   ## to gain experience to create function
```

```python
whr_pd.loc[whr_pd["Country or region"]=="Finland","Score"]
```

```python
## so by using above logic we can create function by generalizing it
```

```python
def country_score(country):
    score=whr_pd.loc[whr_pd["Country or region"]==country,"Score"]
    return score
```

```python
country_score("Netherlands")
```

```python
country_score("Denmark")
```
