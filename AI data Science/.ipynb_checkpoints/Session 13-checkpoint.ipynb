```python
monthly_visitors = {} # {"January 2015": 400, "February 2015":289}

for row in potus:
    appt = row[2] # data type -> datetime:
    month_yr = appt.strftime("%B %Y")
    if month_yr not in monthly_visitors:
        monthly_visitors[month_yr] = 1
    
    monthly_visitors[month_yr] += 1
```
