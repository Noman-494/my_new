```python
with open("t20_matches.csv") as f:
    t20_matches=list(csv.reader(f))
```

```python
header=t20_matches[0]
data=t20_matches[1:]
```

```python
Serious_counts={}
for i in data:
    series=i[1]
    if series not in Serious_counts:
        Serious_counts[series] =1
    else:
        Serious_counts[series] +=1
    
```
