```python
def frq(data, index):
    
    freq_count = {}

    for i in data:
        col = i[index]
        if col in freq_count:
            freq_count[col] += 1
        else:
            freq_count[col] = 1
            
    return freq_count        
```

```python
def read_csv(file_name, header = True):
    if header:
        f = open(file_name, encoding = "utf-8")
        read = csv.reader(f)
        data = list(read)
        headers = data[0]
        data = data[1:]
        return headers, data
    else:
        f = open(file_name, encoding = "utf-8")
        read = csv.reader(f)
        data = list(read)
        return data
```
