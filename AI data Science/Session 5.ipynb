```python
Count_={"1000 to 10000":0,"10001 to 100000":0,"100001 to 1000000":0,"1000000 +":0}
for i in store:
    review=int(i[5])
    if review>=0 and review <=10000:
        Count_["1000 to 10000"] +=1
    elif review>10000 and review <=100000:
        Count_["10001 to 100000"] +=1
    elif review>100000 and review <=1000000:
        Count_["100001 to 1000000"] +=1
    else:
        Count_["1000000 +"] +=1
    
```
