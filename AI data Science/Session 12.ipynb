```python
def clean_date(date):
    bad_char = ["(",")","C","c", ".", " ","s", "'"]
    for i in bad_char:
        date = date.replace(i,"")
    return date
```

```python
def int_date(date):
    if "-" in date:
        d = date.split("-")
        fst = int(d[0])
        scnd = int(d[1])
        date = (fst+scnd)//2
    else:
        date = int(date)
        
    return date
```

```python
type(int_date("1947"))
```

```python
int_date("1947-1999")
```
